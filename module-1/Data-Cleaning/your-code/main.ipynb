#### 3. Create a mysql engine to set the connection to the server. Check the connection details in [this link](https://relational.fit.cvut.cz/dataset/Stats).

```python
engine = create_engine('mysql+pymysql://guest:relational@relational.fit.cvut.cz:3306/stats')
sqlstr = 'select * from users'
```

#### 4. Import the users table.

```python
users_df = pd.read_sql_query(sqlstr,engine)
users_df
```

#### 5. Rename Id column to userId.

```python
users_df = users_df.rename(columns={"Id":"userId"})
users_df
```

#### 6. Import the posts table. 

```python
sqlstr2 = 'select * from posts'
posts_df = pd.read_sql_query(sqlstr2,engine)
posts_df
```

#### 7. Rename Id column to postId and OwnerUserId to userId.

```python
posts_df = posts_df.rename(columns={"Id":"postId","OwnerUserId":"userId"})
posts_df
```

#### 8. Define new dataframes for users and posts with the following selected columns:
**users columns**: userId, Reputation, Views, UpVotes, DownVotes  
**posts columns**: postId, Score, userID, ViewCount, CommentCount

```python
new_users_df = users_df[['userId', 'Reputation', 'Views', 'UpVotes', 'DownVotes']]
new_users_df
```

```python
#Here, I deliberately changed the capital D in 'userId' because it gave me an error
new_posts_df = posts_df[['postId', 'Score', 'userId', 'ViewCount', 'CommentCount']]
new_posts_df
```

#### 9. Merge the new dataframes you have created, of users and posts. 
You will need to make an inner [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
#I want to see what for data do we have
#new_users_df.head(60)
new_posts_df.head(60)
```

```python
#I assume here (for the purpose of this exercice) that both userId columns are refering to the same user in both dataframes

new_users_posts = pd.merge(new_users_df,new_posts_df[['userId','postId','Score','ViewCount','CommentCount']], on = 'userId')
new_users_posts
```

#### 10. How many missing values do you have in your merged dataframe? On which columns?

#### 11. You will need to make something with missing values.  Will you clean or filling them? Explain. 
**Remember** to check the results of your code before going to the next step.

```python
#We do have data != 0 in the column ViewCount representing more than 40% of the data in the column therefeore I will not delete them
new_users_posts[['ViewCount']]=new_users_posts[['ViewCount']].fillna(0)
new_users_posts.isnull().sum()
```
